```python
import numpy as np
import pandas as pd
```

```python
total_data = np.array([])
all_name = np.array([])
for i in range(1,5):
    data = pd.read_excel("../../Downloads/stock0{}.xlsx".format(i))
    data = np.array(data)
    name = data[:,0]
    data = data[:,1:]
    total_data = np.append(total_data, data)
    all_name = np.append(all_name, name)
total_data = total_data.reshape(-1,7)
```

```python
names=list(sorted(set(all_name)))
```

```python
len(names)
```

```python
name_dict={}
for key, value in enumerate(names):
    name_dict[value] = key
```

```python
name_dict
```

```python
new_names=np.array(list(map(lambda all_name: name_dict[all_name], all_name)))
```

```python
new_names
```

```python
full_data = np.c_[new_names, total_data]
```

```python
np.save("processed_stock_data", full_data)
```
